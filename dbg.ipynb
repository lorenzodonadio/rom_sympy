```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt


class IdealHAWT:
    """
    Class for calculating performance parameters of an ideal 3-bladed horizontal axis wind turbine
    with wake rotation, including tip-loss correction and Glauert correction.
    """

    def __init__(self, rotor_radius, air_density=1.225):
        """
        Initialize the wind turbine parameters.

        Parameters:
        -----------
        rotor_radius : float
            Rotor radius in meters
        air_density : float
            Air density in kg/mÂ³ (default: 1.225 kg/mÂ³ at sea level)
        """
        self.R = rotor_radius
        self.rho = air_density
        self.A = np.pi * rotor_radius**2  # Rotor swept area

    def tsr(self, wind_speed, angular_velocity):
        """
        Calculate the tip speed ratio.

        Parameters:
        -----------
        wind_speed : float
            Free stream wind speed in m/s
        angular_velocity : float
            Rotor angular velocity in rad/s

        Returns:
        --------
        float : Tip speed ratio
        """
        return (angular_velocity * self.R) / wind_speed

    def local_speed_ratio(self, tsr, radial_pos):
        """
        Calculate the local speed ratio at a given radial position.

        Parameters:
        -----------
        tsr : float
        radial_pos : float Radial position (0 at hub, R at tip)

        Returns:
        --------
        float : Local speed ratio
        """
        return tsr * (radial_pos / self.R)

    def prandtl_tip_loss_factor(self, radial_pos, axial_induction, tsr):
        """
        Calculate Prandtl tip loss correction factor for 3 blades.

        Parameters:
        -----------
        radial_pos : float
            Radial position
        axial_induction : float
            Axial induction factor
        tsr : float
            Tip speed ratio

        Returns:
        --------
        float : Tip loss factor F
        """
        if radial_pos >= self.R:
            return 0.0

        # Local speed ratio at current radial position
        lambda_r = self.local_speed_ratio(tsr, radial_pos)

        # Avoid division by zero
        if lambda_r == 0 or radial_pos == 0:
            return 1.0

        # Calculate flow angle phi
        # if (1 - axial_induction) * lambda_r == 0:
        #     phi = np.pi / 2
        # else:
        #     phi = np.arctan(1 / ((1 - axial_induction) * lambda_r))
        phi = np.arctan(1 / ((1 - axial_induction) * lambda_r))

        # Calculate the argument for the exponential - tip loss
        f_tip = (3 / 2) * ((self.R - radial_pos) / (radial_pos * np.sin(phi)))
        exp_tip = np.exp(-np.clip(f_tip, -100, 100))
        F_tip = (2 / np.pi) * np.arccos(np.clip(exp_tip, -1.0, 1.0))

        # Hub loss factor
        hub_radius = 0.1 * self.R  # Assume hub at 10% of radius
        if radial_pos <= hub_radius:
            f_hub = (3 / 2) * ((radial_pos - hub_radius) / (radial_pos * np.sin(phi)))
            exp_hub = np.exp(-np.clip(f_hub, -100, 100))
            F_hub = (2 / np.pi) * np.arccos(np.clip(exp_hub, -1.0, 1.0))
        else:
            F_hub = 1.0

        return F_tip * F_hub

    def thrust_coefficient_momentum(self, axial_induction, tip_loss_factor=1.0):
        """
        Calculate thrust coefficient from momentum theory.

        Parameters:
        -----------
        axial_induction : float
            Axial induction factor
        tip_loss_factor : float
            Prandtl tip loss factor

        Returns:
        --------
        float : Thrust coefficient C_T
        """
        return 4 * axial_induction * (1 - axial_induction) * tip_loss_factor

    def glauert_correction(
        self, axial_induction, tip_loss_factor=1.0, max_iterations=10, tolerance=1e-6
    ):
        """
        Apply Glauert correction for high axial induction factors (a > 0.4).
        Uses iterative solution to handle the circular dependency.

        Parameters:
        -----------
        axial_induction : float
            Axial induction factor
        tip_loss_factor : float
            Prandtl tip loss factor
        max_iterations : int
            Maximum number of iterations
        tolerance : float
            Convergence tolerance

        Returns:
        --------
        float : Corrected axial induction factor
        """
        if axial_induction <= 0.4:
            return axial_induction

        # Initial guess
        a_current = axial_induction

        for i in range(max_iterations):
            # Calculate thrust coefficient using current axial induction
            CT = self.thrust_coefficient_momentum(a_current, tip_loss_factor)

            # Apply Glauert correction formula
            discriminant = 0.0203 - 0.6427 * (0.889 - CT)
            # print("Discriminant: ", discriminant)

            if discriminant < 0:
                # If discriminant is negative, use fallback
                a_new = 0.4 + (axial_induction - 0.4) * 0.5  # Dampened correction
            else:
                a_new = (0.143 + np.sqrt(discriminant)) / tip_loss_factor
                # print(a_new)

            # Check for convergence
            if abs(a_new - a_current) < tolerance:
                return np.clip(a_new, 0.0, 1.0)

            a_current = a_new

        return np.clip(a_current, 0.0, 1.0)

    def thrust_coefficient_glauert(self, axial_induction, tip_loss_factor=1.0):
        """
        Calculate thrust coefficient with Glauert correction.

        Parameters:
        -----------
        axial_induction : float
            Axial induction factor
        tip_loss_factor : float
            Prandtl tip loss factor

        Returns:
        --------
        float : Thrust coefficient C_T
        """
        if axial_induction <= 0.4:
            # Standard momentum theory
            return self.thrust_coefficient_momentum(axial_induction, tip_loss_factor)
        else:
            return ((axial_induction * tip_loss_factor - 0.143) ** 2 + 0.55) / 0.6427

    def axial_induction_from_lambda_r(self, lambda_r, radial_pos, tsr, use_corrections=True):
        """
        Calculate axial induction factor for maximum power at given local speed ratio.

        Parameters:
        -----------
        lambda_r : float
            Local speed ratio
        radial_pos : float
            Radial position for tip loss calculation
        tsr : float
            Tip speed ratio for tip loss calculation
        use_corrections : bool
            Whether to apply tip loss and Glauert corrections

        Returns:
        --------
        float : Axial induction factor a
        """
        if lambda_r == 0:
            return 0.25

        # Define the equation to solve
        def equation(a):
            return (1 - a) * (4 * a - 1) ** 2 - lambda_r**2 * (1 - 3 * a)

        # Solve for a in the range [0.25, 0.333]
        a_initial = 0.3
        try:
            a_solution = fsolve(equation, a_initial)[0]
            # print(a_solution)
            a_solution = np.clip(a_solution, 0.25, 1 / 3)
        except:
            a_solution = 1 / 3

        # Apply corrections if requested
        if use_corrections:
            # Get tip loss factor
            F = self.prandtl_tip_loss_factor(radial_pos, a_solution, tsr)
            # Apply Glauert correction
            a_solution = self.glauert_correction(a_solution, F)

        return np.clip(a_solution, 0.0, 1.0)

    def angular_induction_factor(self, axial_induction, radial_pos, tsr, use_corrections=True):
        """
        Calculate angular induction factor from axial induction factor.

        Parameters:
        -----------
        axial_induction : float
            Axial induction factor a
        radial_pos : float
            Radial position for tip loss calculation
        tsr : float
            Tip speed ratio for tip loss calculation
        use_corrections : bool
            Whether to apply tip loss correction

        Returns:
        --------
        float : Angular induction factor a'
        """
        if axial_induction <= 0.25 or abs(4 * axial_induction - 1) < 1e-10:
            return 0.0

        a_prime = (1 - 3 * axial_induction) / (4 * axial_induction - 1)

        # Apply tip loss correction to angular induction factor
        if use_corrections:
            F = self.prandtl_tip_loss_factor(radial_pos, axial_induction, tsr)
            if F > 1e-10:  # Avoid division by zero
                a_prime = a_prime / F

        return max(a_prime, 0.0)

    def power_coefficient(self, tsr, n_points=100, use_corrections=True):
        """
        Calculate the power coefficient by integrating over the rotor.

        Parameters:
        -----------
        tsr : float
            Tip speed ratio
        n_points : int
            Number of integration points
        use_corrections : bool
            Whether to apply corrections

        Returns:
        --------
        float : Power coefficient C_P
        """
        if tsr <= 0:
            return 0.0

        radial_poss = np.linspace(0.1 * self.R, 0.95 * self.R, n_points)
        d_lambda_r = tsr / n_points

        integral = 0.0

        for r in radial_poss:
            lambda_r = self.local_speed_ratio(tsr, r)

            if lambda_r <= 0:
                continue

            a = self.axial_induction_from_lambda_r(lambda_r, r, tsr, use_corrections)
            a_prime = self.angular_induction_factor(a, r, tsr, use_corrections)

            # Get tip loss factor
            F = self.prandtl_tip_loss_factor(r, a, tsr) if use_corrections else 1.0

            integral += F * a_prime * (1 - a) * lambda_r**3 * d_lambda_r

        cp = (8 / tsr**2) * integral
        return min(cp, 0.5926)

    def annular_torque(self, wind_speed, angular_velocity, radial_pos, dr, use_corrections=True):
        """
        Calculate torque on an annular element at radius r with thickness dr.

        Parameters:
        -----------
        wind_speed : float
            Free stream wind speed in m/s
        angular_velocity : float
            Rotor angular velocity in rad/s
        radial_pos : float
            Radial position of the annular element
        dr : float
            Thickness of the annular element
        use_corrections : bool
            Whether to apply corrections

        Returns:
        --------
        float : Torque on annular element in NÂ·m
        """
        lambda_tsr = self.tsr(wind_speed, angular_velocity)
        lambda_r = self.local_speed_ratio(lambda_tsr, radial_pos)

        a = self.axial_induction_from_lambda_r(lambda_r, radial_pos, lambda_tsr, use_corrections)
        a_prime = self.angular_induction_factor(a, radial_pos, lambda_tsr, use_corrections)

        # Calculate tip loss factor
        F = self.prandtl_tip_loss_factor(radial_pos, a, lambda_tsr) if use_corrections else 1.0

        # Calculate annular area and torque
        dA = 2 * np.pi * radial_pos * dr
        Cq = 4 * F * a_prime * (1 - a)
        dQ = 0.5 * Cq * self.rho * wind_speed * angular_velocity * radial_pos**2 * dA
        return dQ

    def total_torque(self, wind_speed, angular_velocity, n_annuli=100, use_corrections=True):
        """
        Calculate total torque on the rotor by integrating annular torques.

        Parameters:
        -----------
        wind_speed : float
            Free stream wind speed in m/s
        angular_velocity : float
            Rotor angular velocity in rad/s
        n_annuli : int
            Number of annular elements for integration
        use_corrections : bool
            Whether to apply corrections

        Returns:
        --------
        float : Total torque in NÂ·m
        """
        dr = self.R / n_annuli
        total_torque = 0.0

        for i in range(n_annuli):
            r = (i + 0.5) * dr
            if r < 0.1 * self.R:  # Skip very small radii near hub
                continue
            dQ = self.annular_torque(wind_speed, angular_velocity, r, dr, use_corrections)
            if np.isfinite(dQ):  # Only add finite values
                total_torque += dQ

        return total_torque

    def thrust_coefficient(self, tsr, n_points=100, use_corrections=True):
        """
        Calculate thrust coefficient C_T.

        Parameters:
        -----------
        tsr : float
            Tip speed ratio
        n_points : int
            Number of integration points
        use_corrections : bool
            Whether to apply corrections

        Returns:
        --------
        float : Thrust coefficient C_T
        """
        if tsr <= 0:
            return 0.0

        radial_poss = np.linspace(0.1 * self.R, 0.95 * self.R, n_points)

        thrust = 0.0
        reference_dynamic_pressure = 0.5 * self.rho * (self.R * tsr) ** 2
        dr = self.R / n_points
        for r in radial_poss:
            dA = 2 * np.pi * r * dr
            lambda_r = self.local_speed_ratio(tsr, r)

            if lambda_r <= 0:
                continue

            a = self.axial_induction_from_lambda_r(lambda_r, r, tsr, use_corrections)
            F = self.prandtl_tip_loss_factor(r, a, tsr) if use_corrections else 1.0

            # Annular thrust
            dT = 0.5 * self.rho * (self.R * tsr) ** 2 * self.thrust_coefficient_glauert(a, F)

            if np.isfinite(dT):
                thrust += dT * dA

        return thrust / (reference_dynamic_pressure * self.A)

    def power_output(
        self, wind_speed, angular_velocity, mechanical_efficiency=0.95, use_corrections=True
    ):
        """
        Calculate mechanical power output.

        Parameters:
        -----------
        wind_speed : float
            Free stream wind speed in m/s
        angular_velocity : float
            Rotor angular velocity in rad/s
        mechanical_efficiency : float
            Mechanical/electrical efficiency (default: 0.95)
        use_corrections : bool
            Whether to apply corrections

        Returns:
        --------
        float : Power output in Watts
        """
        lambda_tsr = self.tsr(wind_speed, angular_velocity)
        cp = self.power_coefficient(lambda_tsr, use_corrections=use_corrections)
        if not np.isfinite(cp):
            cp = 0.0
        power_wind = 0.5 * self.rho * self.A * wind_speed**3
        return mechanical_efficiency * cp * power_wind


def calculate_torque(rotor_radius, wind_speed, angular_velocity, use_corrections=True):
    """
    Simple function to calculate torque for given conditions.

    Parameters:
    -----------
    rotor_radius : float
        Rotor radius in meters
    wind_speed : float
        Wind speed in m/s
    angular_velocity : float
        Angular velocity in rad/s
    use_corrections : bool
        Whether to use tip loss and Glauert corrections

    Returns:
    --------
    float : Torque in NÂ·m
    """
    turbine = IdealHAWT(rotor_radius)
    torque = turbine.total_torque(wind_speed, angular_velocity, use_corrections=use_corrections)
    return torque if np.isfinite(torque) else 0.0


def calculate_power_coefficient(rotor_radius, tsr, use_corrections=True):
    """
    Simple function to calculate power coefficient.

    Parameters:
    -----------
    rotor_radius : float
        Rotor radius in meters
    tsr : float
        Tip speed ratio
    use_corrections : bool
        Whether to use corrections

    Returns:
    --------
    float : Power coefficient C_P
    """
    turbine = IdealHAWT(rotor_radius)
    cp = turbine.power_coefficient(tsr, use_corrections=use_corrections)
    return cp if np.isfinite(cp) else 0.0


def plot_tip_loss_distribution(tsr=7.5):
    """
    Plot tip loss factor distribution across the rotor
    """
    turbine = IdealHAWT(rotor_radius=1.0)

    radial_poss = np.linspace(0.1, 1.0, 100)
    tip_loss_factors = []

    for r in radial_poss:
        # Calculate tip loss factor for a typical axial induction
        F = turbine.prandtl_tip_loss_factor(r, 1 / 3, tsr)
        tip_loss_factors.append(F)

    plt.figure(figsize=(10, 6))
    plt.plot(radial_poss, tip_loss_factors, "b-", linewidth=2)
    plt.xlabel("Normalized Radial Position (r/R)")
    plt.ylabel("Tip Loss Factor, F")
    plt.title(f"Prandtl Tip Loss Factor Distribution (Î» = {tsr})")
    plt.grid(True, alpha=0.3)
    plt.ylim(0, 1.1)
    plt.show()


def plot_thrust_coefficient_comparison():
    """
    Plot thrust coefficient with and without Glauert correction
    """
    turbine = IdealHAWT(rotor_radius=1.0)

    # Simulate axial induction factors
    a_values = np.linspace(0.0, 0.8, 100)
    ct_uncorrected = 4 * a_values * (1 - a_values)  # Standard momentum theory
    ct_corrected = []

    for a in a_values:
        # Apply Glauert correction
        ct_corrected.append(turbine.thrust_coefficient_glauert(a, 1))

    plt.figure(figsize=(10, 6))
    plt.plot(a_values, ct_uncorrected, "b-", linewidth=2, label="Standard momentum theory")
    plt.plot(a_values, ct_corrected, "r-", linewidth=2, label="With Glauert correction")
    plt.axvline(x=0.4, color="k", linestyle="--", alpha=0.5, label="Glauert transition (a=0.4)")
    plt.xlabel("Axial Induction Factor, a")
    plt.ylabel("Thrust Coefficient, C_T")
    plt.title(
        "Thrust Coefficient vs Axial Induction Factor (No Tip Loss)\n(Glauert Correction Effect)"
    )
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.ylim(0, 1.5)
    plt.show()


def main():
    import numpy as np

    """
    Demonstration of the 3-bladed wind turbine analysis.
    """
    # Initialize turbine
    turbine = IdealHAWT(rotor_radius=50.0)

    print("3-BLADED WIND TURBINE ANALYSIS")
    print("=" * 50)

    # Example calculations
    wind_speed = 12.0
    angular_velocity = 1.8

    # Calculate performance metrics
    tsr = turbine.tsr(wind_speed, angular_velocity)
    cp = turbine.power_coefficient(tsr, use_corrections=True)
    ct = turbine.thrust_coefficient(tsr, use_corrections=True)
    torque = turbine.total_torque(wind_speed, angular_velocity, use_corrections=True)
    power = turbine.power_output(wind_speed, angular_velocity, use_corrections=True)

    print(f"Rotor radius: {turbine.R} m")
    print(f"Wind speed: {wind_speed} m/s")
    print(f"Angular velocity: {angular_velocity} rad/s")
    print(f"Tip speed ratio: {tsr:.3f}")
    print(f"Power coefficient C_P: {cp:.4f}")
    print(f"Thrust coefficient C_T: {ct:.4f}")
    print(f"Torque: {torque:,.0f} NÂ·m")
    print(f"Power output: {power/1e6:.2f} MW")
    print(f"Available wind power: {0.5*turbine.rho*turbine.A*wind_speed**3/1e6:.2f} MW")

    print("\n" + "=" * 50)
    print("TORQUE CALCULATIONS FOR DIFFERENT CONDITIONS")
    print("=" * 50)

    # Torque for different conditions
    # wind_speeds = [8, 10, 12]
    wind_speeds = np.linspace(1, 25, 50)
    angular_velocities = [0.5, 1, 1.2, 1.5, 1.8]

    print(f"{'Wind Speed':<12} {'Angular Velocity':<18} {'TSR':<8} {'Torque':<15} {'Ct':<8}")
    print(f"{'(m/s)':<12} {'(rad/s)':<18} {'':<8} {'(kNÂ·m)':<15} - ")
    print("-" * 55)

    # for ws in wind_speeds:
    #     for av in angular_velocities:
    for av in angular_velocities:
        for ws in wind_speeds:
            torque_val = calculate_torque(50.0, ws, av, use_corrections=True) / 1000

            tsr_val = turbine.tsr(ws, av)
            ct = turbine.thrust_coefficient(tsr_val, use_corrections=True)
            print(f"{ws:<12.1f} {av:<18.1f} {tsr_val:<8.2f} {torque_val:<15.1f} {ct:8.2f}")

        # Create plots
    print("\nGenerating plots...")
    plot_tip_loss_distribution()
    plot_thrust_coefficient_comparison()
```
